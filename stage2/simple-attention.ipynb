# Attention 

Simple self-attention without trainable weights

```python
import torch

inputs = torch.tensor(
  [[0.43, 0.15, 0.89], # Your     (x^1)
   [0.55, 0.87, 0.66], # journey  (x^2)
   [0.57, 0.85, 0.64], # starts   (x^3)
   [0.22, 0.58, 0.33], # with     (x^4)
   [0.77, 0.25, 0.10], # one      (x^5)
   [0.05, 0.80, 0.55]] # step     (x^6)
)
```

calculate attention weights for input x(1) 

```python
attention_weights_1 = torch.zeros(inputs.shape[0])
query = inputs[1]

for i, vector in enumerate(inputs):
    attention_weights_1[i] = torch.dot(query, vector)

attention_weights_1
```

```python
norm_attention_weights_1 = torch.softmax(attention_weights_1, dim=0)
norm_attention_weights_1
```

```python
torch.sum(norm_attention_weights_1)
```

weighted sum (norm_attention_weights_1 * vector) -> context vector z(1)

```python
context_vector = 0.

for i, vector in enumerate(inputs):
    context_vector += norm_attention_weights_1[i] * vector

context_vector
```
